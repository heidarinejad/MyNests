```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline
```

```python
def findfile(keyword,root):
    filelist=[]
    find_file_list=[]
    for root,dirs,files in os.walk(root):
        for name in files:                
            fitfile=filelist.append(os.path.join(root, name))
            #print(fitfile)
            #print(os.path.join(root, name))
    #print(filelist)
    print('...........................................')
    for i in filelist:            
        if os.path.isfile(i):
            
            #print(i)
            if keyword in os.path.split(i)[1]:
                find_file_list.append(i)
            #else:
                #print('......no keyword!')
    return find_file_list
root=os.path.abspath('.')
results_filelist=[]

results_filelist.append(findfile('layer_gr',root))
print (results_filelist[0])
```

```python
for i in range (len(results_filelist[0])):
    file=results_filelist[0][i]
    print (file)
    spike_list=pd.read_csv(file,header=None, delimiter=r"\s+").values
    if i ==0:
        sd_data=spike_list
    else:
        sd_data=np.concatenate((sd_data,spike_list),axis=0)
    print (spike_list.shape)
    print (sd_data.shape)
```

```python
plt.figure(figsize=(15,5))
plt.plot(sd_data[:, 1], sd_data[:, 0], '.')
plt.show()
```
